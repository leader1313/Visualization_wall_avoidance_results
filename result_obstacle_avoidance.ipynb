```python
class VideoMaker():
    def __init__(self, video_name='video_name'):
        yd_string, time_string = now_stamp()
        dir_ID = video_name + '/' + yd_string
        self.figure_dir = 'Figures/' + dir_ID + '/' + time_string + '/'
        self.video_dir = 'Videos/' + dir_ID + '/'
        dirs = {'fig': self.figure_dir, 'video': self.video_dir}
        for key in dirs:
            if not os.path.exists(dirs[key]):
                os.makedirs(dirs[key])

        self.file_name = time_string

    # snapshots from learning process -------
    def get_plot(self, step, dataset, color='navy'):
        xx = torch.linspace(1, step + 1, steps=step + 1)
        yy = dataset[:step + 1]
        plt.plot(xx, yy, color=color, linewidth=5)

    def get_figure(self, step, dataset, init=None, color='tomato'):
        plt.style.use("ggplot")

        if init is not None:
            init()

        self.get_plot(step=step,dataset=dataset,color=color)
        
        plt.xlim(0, 220)
        plt.ylim(0, 1.0)
        plt.xticks(fontsize=10)
        plt.yticks(fontsize=10)

        file_name = self.figure_dir + 'frame' + str(step)
        # plt.savefig(file_name)
        # plt.clf()
        plt.show()
```

```python
MGP = torch.load('Data/Disturbances/144719.pickle')
MHGP = torch.load('Data/Disturbances/141223.pickle')
```

```python
videoMaker = VideoMaker(video_name='test')
init_plt = videoMaker.get_plot(step=100, dataset=MGP, color='navy')
for s in range(3):
    videoMaker.get_figure(step=100, dataset=MHGP, init=init_plt, color='tomato')
```
